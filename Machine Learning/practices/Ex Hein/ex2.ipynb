```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('./data/fifa_data.csv', sep=',', index_col=False)
```

```python
df
```

```python
overall = df['Overall']
bins = range(40, 101, 5)
```

```python
plt.hist(overall, bins = bins)
plt.title("Overall skill level of the fifa players")
plt.xlabel("Overall")
plt.ylabel('Number of players')
plt.show()
```

```python
left = df[df['Preferred Foot'] == 'Left'].shape[0]
right = df[df['Preferred Foot'] == 'Right'].shape[0]

votes = ['Left', 'Right']
pref = [left, right]

plt.pie(pref, labels = votes, autopct='%.2f%%')
plt.show()
```

```python
df['Weight'] = df['Weight'].str.replace('lbs', '', regex=False).astype(float)
```

```python
df.rename(columns={'Weight':'Weight(lbs)'}, inplace=True)
```

```python
df['Weight(kg)'] = df['Weight(lbs)'] * 0.4536
```

```python
light = df.loc[df['Weight(kg)'] < 60].count()[0]
light_medium = df[(df['Weight(kg)'] >= 60) & (df['Weight(kg)'] < 70)].count()[0]
medium = df[(df['Weight(kg)'] >= 70) & (df['Weight(kg)'] < 80)].count()[0]
medium_heavy = df[(df['Weight(kg)'] >= 80) & (df['Weight(kg)'] < 90)].count()[0]
heavy = df[df['Weight(kg)'] >= 90].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
label = ['under 60', '60-70', '70-80', '80-90', 'over 90']
explode = (.4,.2,0,0,.4)

plt.pie(weights, labels= label, explode=explode, autopct='.%2f%%')
```

```python
df = pd.read_csv("./data/gas_price.csv")
```

```python
df.dropna(inplace=True)
x = df['Year']
```

```python
plt.xticks(x[::3].astype(int))
plt.plot(x, df['USA'], lw = 2, marker = 'o')
plt.plot(x, df['Canada'], lw = 2, marker = 'o')
plt.plot(x, df['Australia'], lw = 2, marker = 'o')
plt.plot(x, df['South Korea'], lw = 2, marker = 'o')

plt.title('Gas prices for particular countries')
plt.xlabel('Year')
plt.ylabel('Gas per Gallons')
```

```python
plt.figure(figsize=(14, 6))

for country in df.columns[1:]: 
    plt.plot(df['Year'], df[country], label=country, marker='o', lw=2)

plt.title('Gas Prices by Country Over Time')
plt.xlabel('Year')
plt.ylabel('Gas Price (per Gallon)')
plt.xticks(df['Year'][::3])  
plt.legend()
plt.show()
```
