```python
import pandas as pd
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data",delim_whitespace=True)

df
```

```python
df.columns = ["mpg", "cylinders", "displacement", "horsepower", "weight","acceleration", "model_year", "origin", "car_name"]
```

```python
df
```

```python
df.info()
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'] , errors="coerce")
```

```python
df.info()
```

```python
df.dropna()
```

```python

df.loc[(df['cylinders'] == 8) & (df['model_year'] == 70), 'bestsellers'] = 1

df.loc[(df['cylinders'] == 6) & (df['model_year'] == 80), 'bestsellers'] = 2
```
