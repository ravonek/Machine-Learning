```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('price_paid_records_small.csv')
```

```python
prices = prices.dropna()
```

```python
postcodes = pd.read_csv('postcodes.csv')
```

```python
postcodes = postcodes.dropna()
```

```python
postcodes['town'] = postcodes['town'].str.upper()
```

```python
postcodes.drop_duplicates(subset=['town'], inplace=True)
```

```python
joined = prices.set_index('Town/City').join(postcodes.set_index('town'), how='inner')
```

```python
east = joined.loc[joined['longitude'] > 0]
west = joined.loc[joined['longitude'] <= 0]
```

```python
plt.scatter(east.longitude, east.latitude, c='g', s=20)
```

```python
plt.scatter(west.longitude, west.latitude, c='r', s=20)
```

```python
plt.scatter(east.longitude, east.latitude, c='g', s=20)
plt.scatter(west.longitude, west.latitude, c='r', s=20)
```

```python
plt.scatter(east.longitude, east.latitude, c='g', s=50)
plt.scatter(west.longitude, west.latitude, c='r', s=50)
plt.scatter(0,51.48092965544458, c='k', s=200)

plt.title('Sales in East/West Greenwich')
plt.xlabel('longitude', fontsize=12)
plt.ylabel('latitude', fontsize=12)
```
